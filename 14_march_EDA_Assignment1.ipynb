```python
high_bp_prevalence = df.groupby('sex')['high_blood_pressure'].mean()
print(high_bp_prevalence)
```

```python
plt.figure(figsize=(8, 6))
high_bp_prevalence.plot(kind='bar', color=['blue', 'red'])
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Sex')
plt.ylabel('Prevalence of High Blood Pressure')
plt.xticks([0, 1], ['Female', 'Male'], rotation=0)
plt.ylim(0, 1)  # Ensure y-axis starts from 0 and ends at 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# What is the relationship between smoking habits and the occurrence of heart failure
death_event_smokers = df[df['DEATH_EVENT'] == 1]['smoking'].mean()

# Calculate the proportion of smokers among patients who did not experience a death event (heart failure)
no_death_event_smokers = df[df['DEATH_EVENT'] == 0]['smoking'].mean()

# Plotting the results
plt.figure(figsize=(8, 6))
plt.bar(['Heart Failure', 'No Heart Failure'], [death_event_smokers, no_death_event_smokers], color=['red', 'green'])
plt.title('Smoking Habits and Occurrence of Heart Failure')
plt.ylabel('Proportion of Smokers')
plt.ylim(0, 1)  # Ensure y-axis starts from 0 and ends at 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Are there any noticeable patterns in the distribution of death events across different age groups
plt.figure(figsize=(10, 6))
plt.hist(df[df['DEATH_EVENT'] == 1]['age'], bins=20, color='red', alpha=0.7, label='Death Event')
plt.hist(df[df['DEATH_EVENT'] == 0]['age'], bins=20, color='blue', alpha=0.7, label='No Death Event')
plt.title('Distribution of Death Events Across Age Groups')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
survived = df[df['DEATH_EVENT'] == 0]['serum_creatinine']
not_survived = df[df['DEATH_EVENT'] == 1]['serum_creatinine']

plt.figure(figsize=(8, 6))
plt.boxplot([survived, not_survived], labels=['Survived', 'Not Survived'])
plt.title('Serum Creatinine Levels by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Serum Creatinine Level')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
